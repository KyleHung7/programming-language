```python
import pandas as pd
import random

# Define food portions and their respective calorie values (estimated based on average serving sizes)
scores = {
    "class1": {
        "Amy": 65,
        "Billy": 80,
        "John": 73,
        "Tommy": 90,
        "Jeff": 100,
        "Will": 50
    },
    "class2": {
        "Tom": 70,
        "Jesse": 80,
        "Wish": 45,
        "Felix": 100,

    }
}
```

```python
scores['class2'].keys()
```

```python
import pandas as pd
df = pd.DataFrame(list(scores['class2'].keys()), columns=['class2'])
```

```python
df
```

```python
# 轉換為長格式
data_all = []
for category, items in scores.items():
    for food, calories in items.items():
        data_all.append([category, food, calories])

# 建立 DataFrame
df_all = pd.DataFrame(data_all, columns=['Classes', 'Students', 'Scores'])
```
