```python
# 設定每年 df 的標題列當作欄位名稱
df_2019.columns = df_2019.iloc[0]
df_2020.columns = df_2020.iloc[0]
df_2021.columns = df_2021.iloc[0]
df_2022.columns = df_2022.iloc[0]
df_2023.columns = df_2023.iloc[0]

# 去掉標題列本身（第一列）
df_2019 = df_2019.iloc[1:]
df_2020 = df_2020.iloc[1:]
df_2021 = df_2021.iloc[1:]
df_2022 = df_2022.iloc[1:]
df_2023 = df_2023.iloc[1:]

# 對齊欄位名稱，避免有空白或換行問題
dfs = [df_2019, df_2020, df_2021, df_2022, df_2023]
for i in range(len(dfs)):
    dfs[i].columns = dfs[i].columns.str.strip()

# 只保留需要的欄位，並重新命名後合併
def extract_fields(df, year_from, year_to):
    return df[[
        '公司代碼', '公司名稱', '類別代碼', '類別名稱',
        f'{year_from}年營業收入',
        f'{year_from}-{year_to}年營業收入成長率',
        f'{year_to}年營業收入',
        f'{year_from}年營業利益',
        f'{year_to}年營業利益',
        f'{year_from}年營業外收入及支出合計',
        f'{year_to}年營業外收入及支出合計',
        f'{year_from}年稅後淨利(淨損)',
        f'{year_to}年稅後淨利(淨損)',
        f'{year_from}-{year_to}年稅後淨利成長率(%)',
        f'{year_to}年股本',
        f'{year_from}年每股盈餘(元)',
        f'{year_to}年每股盈餘(元)',
        f'{year_to}年每股淨值(NT$)',
        f'{year_to}年淨值佔總資產(%)',
        f'{year_to}年流動比率',
        f'{year_to}年速動比率',
        f'{year_from}年稅前淨利(淨損)',
        f'{year_to}年稅前淨利(淨損)',
        f'{year_from}-{year_to}年稅前淨利(淨損)成長率'
    ]].copy()

df_merge = extract_fields(df_2019, 2018, 2019)
df_merge = df_merge.merge(extract_fields(df_2020, 2019, 2020), on=['公司代碼', '公司名稱', '類別代碼', '類別名稱'], how='outer')
df_merge = df_merge.merge(extract_fields(df_2021, 2020, 2021), on=['公司代碼', '公司名稱', '類別代碼', '類別名稱'], how='outer')
df_merge = df_merge.merge(extract_fields(df_2022, 2021, 2022), on=['公司代碼', '公司名稱', '類別代碼', '類別名稱'], how='outer')
df_merge = df_merge.merge(extract_fields(df_2023, 2022, 2023), on=['公司代碼', '公司名稱', '類別代碼', '類別名稱'], how='outer')

# 檢查結果
df_merge.head()


```

```python
import jieba
from snownlp import SnowNLP
import pandas as pd
from collections import Counter
import re
import plotly.express as px
from wordcloud import WordCloud


analysis_text = response.text

# ======== 分析與詞頻統計 ========

# 1. 只保留中文
def extract_chinese(text):
    return ''.join(re.findall(r'[\u4e00-\u9fa5]', text))

# 2. 自定義詞彙加入斷詞字典（可再擴充）
custom_dict = ['營收', '淨利', '每股盈餘', 'EPS', '資產', '負債', '流動比率', '速動比率', '淨值']
for word in custom_dict:
    jieba.add_word(word, freq=10000)
    jieba.suggest_freq(word, tune=True)

# 3. 放入 DataFrame
df = pd.DataFrame({'Content': [analysis_text]})
df['Chinese_Content'] = df['Content'].apply(extract_chinese)

# 4. 定義三種斷詞模式
def tokenize_modes(text):
    return {
        '精確模式': list(jieba.cut(text, HMM=True)),
        '全模式': list(jieba.cut(text, cut_all=True)),
        '搜尋模式': list(jieba.cut_for_search(text))
    }

# 5. 詞頻統計
mode_word_freq = {}
for mode, tokens in tokenize_modes(df['Chinese_Content'][0]).items():
    filtered = [word for word in tokens if len(word) > 1]
    counter = Counter(filtered)
    word_freq_df = pd.DataFrame(counter.items(), columns=['Word', 'Frequency'])
    word_freq_df = word_freq_df.sort_values(by='Frequency', ascending=False).reset_index(drop=True)
    mode_word_freq[mode] = word_freq_df

# ======== 可視化（以精確模式為例） ========
# 條形圖
fig = px.bar(mode_word_freq['精確模式'].head(30), x='Word', y='Frequency',
             title='關鍵字出現次數統計（精確模式 前 30 名）',
             text='Frequency')
fig.update_layout(template='plotly_white')
fig.show()

# 文字雲
top_words = mode_word_freq['精確模式'].head(300)
wordcloud = WordCloud(
    width=800,
    height=400,
    background_color='white',
    font_path='/usr/share/fonts/truetype/wqy/wqy-zenhei.ttc'
)
wordcloud.generate_from_frequencies(dict(zip(top_words['Word'], top_words['Frequency'])))
wordcloud_img = wordcloud.to_array()

fig = px.imshow(wordcloud_img)
fig.update_layout(
    title="文字雲（精確模式）",
    xaxis=dict(showgrid=False, showticklabels=False, zeroline=False),
    yaxis=dict(showgrid=False, showticklabels=False, zeroline=False),
    coloraxis_showscale=False
)
fig.show()

# 最後查看詞頻結果表
word_freq_df.head(50)
```

#### 6.機器學習分析

```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import silhouette_score

# ✅ Step 1: 清理與前處理資料
df = df_merge.copy()

# 將類別代碼保留做 One-Hot Encoding 用
df["類別代碼"] = df["類別代碼"].astype(str)

# 將非數值欄位移除（只保留數值 + 類別編碼）
non_numeric_cols = ["公司代碼", "公司名稱", "類別名稱"]
df_cleaned = df.drop(columns=non_numeric_cols).copy()

# One-Hot Encoding on 類別代碼
ohe = OneHotEncoder(sparse_output=False, handle_unknown="ignore")
category_encoded = ohe.fit_transform(df[["類別代碼"]])
category_df = pd.DataFrame(category_encoded, columns=ohe.get_feature_names_out(["類別代碼"]))

# 合併 One-Hot 結果
df_cleaned = df_cleaned.drop(columns=["類別代碼"]).reset_index(drop=True)
df_cleaned = pd.concat([df_cleaned, category_df], axis=1)

# 去除千分位符號，轉換成數值格式
df_cleaned = df_cleaned.replace(",", "", regex=True).apply(pd.to_numeric, errors="coerce")

# 去除含 NaN 的列
df_cleaned = df_cleaned.dropna()
df = df.loc[df_cleaned.index].reset_index(drop=True)

# ✅ Step 2: 標準化資料
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_cleaned)

# ✅ Step 3: PCA 降維
pca = PCA(n_components=2)
pca_result = pca.fit_transform(df_scaled)
df["PCA1"] = pca_result[:, 0]
df["PCA2"] = pca_result[:, 1]

# ✅ Step 4: 尋找最佳 K 值
inertia = []
sil_scores = []
K_range = range(2, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(df_scaled)
    inertia.append(kmeans.inertia_)
    sil_scores.append(silhouette_score(df_scaled, kmeans.labels_))

# Elbow Method 圖形
fig_elbow = px.line(
    x=list(K_range),
    y=inertia,
    markers=True,
    title="Elbow Method to Find Best K",
    labels={"x": "Number of Clusters (K)", "y": "Inertia (SSE)"},
)
fig_elbow.show()

# ✅ Step 5: KMeans 分群（假設我們選擇 K=4，可調整）
best_k = 4
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(df_scaled)

# ✅ Step 6: PCA + Cluster 可視化
fig_cluster = px.scatter(
    df, x="PCA1", y="PCA2", color=df["Cluster"].astype(str),
    title="PCA + K-Means Clustering Result",
    labels={"PCA1": "PCA Component 1", "PCA2": "PCA Component 2", "Cluster": "Cluster Group"},
)
fig_cluster.show()

# ✅ Step 7: Heatmap of Feature Correlations
corr_matrix = df_cleaned.corr()
fig_heatmap = go.Figure(data=go.Heatmap(
    z=corr_matrix.values,
    x=corr_matrix.columns,
    y=corr_matrix.index,
    colorscale="viridis"
))
fig_heatmap.update_layout(title="Feature Correlation Heatmap")
fig_heatmap.show()

```

```python
import plotly.graph_objects as go

# Normalize 關鍵財務欄位
from sklearn.preprocessing import MinMaxScaler

features = ["2023年營業收入", "2023年營業利益", "2023年稅後淨利(淨損)", "2023年每股盈餘(元)"]
data = df[features].replace(",", "", regex=True).apply(pd.to_numeric, errors="coerce")
data_scaled = MinMaxScaler().fit_transform(data)

df_scaled = df.copy()
df_scaled[features] = data_scaled

# 計算每個群集的平均特徵值
radar_data = df_scaled.groupby("Cluster")[features].mean().reset_index()

# 繪製雷達圖
fig = go.Figure()
for i, row in radar_data.iterrows():
    fig.add_trace(go.Scatterpolar(
        r=row[features].values,
        theta=features,
        fill='toself',
        name=f"Cluster {int(row['Cluster'])}"
    ))

fig.update_layout(
    polar=dict(radialaxis=dict(visible=True, range=[0, 1])),
    title="各群集財務特徵雷達圖",
    showlegend=True
)
fig.show()
```
